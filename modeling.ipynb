### Used Functions

```python
def format_featnames(data):
    names = []
    featnames = []
    for line in data:
        names = line.split(' ',1)
        string = names[1]
        string = string[:-1]
        names[1] = string
        featnames.append(names[1])
        
    return featnames

def format_feats(data):
    feat = []
    id_list = []
    for line in data:
        feats = line.split()
        id_list.append(feats[0])
        feat.append(feats[1:])
    return id_list, pd.DataFrame(feat).T
```

### Reading all edges datasets and modeling

```python
dfs = []
edge = []
edges = []

for i in range(4):
    dfs.append("dataset/"+ open(str(i) + ".edges", "r"))

for df in dfs:
    for line in df:
        edge = line.split()
        edges.append(edge)
        
```

```python
edges = pd.DataFrame(edges)
edges.columns = edges.iloc[0]
edges = edges.drop(edges.index[0])

edges.head()
```

### Reading all featnames datasets and modeling

```python
dfs = []
names = []
featname = []

for i in range(4):
    dfs.append("dataset/"+ open(str(i) + "featnames", "r"))
    featname.append(format_featnames(dfs[i]))
```

```python
fn0 = pd.DataFrame(featname[0])
fn1 = pd.DataFrame(featname[1])
fn2 = pd.DataFrame(featname[2])
fn3 = pd.DataFrame(featname[3])

featnames = [fn0, fn1, fn2, fn3]
```

### Reading all feats datasets and modeling

```python
dfs = []
feats = []
id_list = []

for i in range(4):
    dfs.append("dataset/"+ open(str(i) + ".feat", "r"))  
```

```python
id_list0, f0 = format_feats(dfs[0])
id_list1, f1 = format_feats(dfs[1])
id_list2, f2 = format_feats(dfs[2])
id_list3, f3 = format_feats(dfs[3])

id_list = [id_list0, id_list1, id_list2, id_list3]
feats = [f0, f1, f2, f3]
```

### Joining featnames and feats

```python
for i in range(4):
    dataset = []
    dataset = feats[i]
    dataset['id'] = featnames[i]
    
    cols = dataset.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    dataset = dataset[cols]
    
    dataset = dataset.T
    dataset.columns = dataset.iloc[0]
    dataset = dataset.drop(dataset.index[0])
    
    dataset['id'] = id_list[i]
    cols = dataset.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    dataset = dataset[cols]
    
    dataset.to_csv("dataset/"+ "dataset"+str(i)+".csv", index=False)
```
